```python
db_path = datasets["sams"]["path"]
conn = sqlite3.connect(db_path)

query = """
SELECT
    barcode,
    aadhar_no,
    academic_year,
    mark_data
FROM students
WHERE module = 'ITI'
"""

students_df = pd.read_sql_query(query, conn)
conn.close()
# Function to parse JSON strings safely
def parse_option_data(val):
    """
    Safely parses a JSON string into a Python object.

    Parameters:
    ----------
    val : str, list, or any
        The value to be parsed. If it's a JSON string, it will be decoded.
        If it's already a list, it is returned unchanged.
        All other types return an empty list.

    Returns:
    -------
    list
        Parsed list from JSON string, or the original list, or an empty list if parsing fails.
    """
    if isinstance(val, str):
        try:
            return json.loads(val)
        except json.JSONDecodeError:
            return []
    elif isinstance(val, list):
        return val
    else:
        return []

students_df['mark_data'] = students_df['mark_data'].apply(parse_option_data)
df_exploded = students_df.explode('mark_data').reset_index(drop=True)
option_details = pd.json_normalize(df_exploded['mark_data'])
iti_df = pd.concat([df_exploded.drop(columns=['mark_data']), option_details], axis=1)

iti_df.head()
```

```python
db_path = datasets["sams"]["path"]
conn = sqlite3.connect(db_path)

query = """
SELECT
    barcode,
    aadhar_no,
    academic_year,
    mark_data
FROM students
WHERE module = 'Diploma'
"""

diploma_df = pd.read_sql_query(query, conn)
conn.close()
# Function to parse JSON strings safely
def parse_mark_data(val):
    """
    Safely parses a JSON string into a Python object.

    Parameters:
    ----------
    val : str, list, or any
        The value to be parsed. If it's a JSON string, it will be decoded.
        If it's already a list, it is returned unchanged.
        All other types return an empty list.

    Returns:
    -------
    list
        Parsed list from JSON string, or the original list, or an empty list if parsing fails.
    """
    if isinstance(val, str):
        try:
            return json.loads(val)
        except json.JSONDecodeError:
            return []
    elif isinstance(val, list):
        return val
    else:
        return []

diploma_df['mark_data'] = diploma_df['mark_data'].apply(parse_mark_data)
diploma_df = diploma_df.explode('mark_data').reset_index(drop=True)
mark_details = pd.json_normalize(diploma_df['mark_data'])
diploma_df = pd.concat([diploma_df.drop(columns=['mark_data']), mark_details], axis=1)

diploma_df.head()
```

```python
datasets = {
    "ITI": iti_df,
    "Diploma": diploma_df,
    "HSS": hss_df,
    "Degree": deg_enrollments
}

results = []

for name, df in datasets.items():
    total = df.shape[0]
    aadhar_present = df["aadhar_no"].notna().sum() if "aadhar_no" in df.columns else None
    aadhar_missing = df["aadhar_no"].isna().sum() if "aadhar_no" in df.columns else None
    
    # handle roll number column variations
    roll_col = None
    for candidate in ["roll_no", "RollNo"]:
        if candidate in df.columns:
            roll_col = candidate
            break
    
    
    if roll_col:
        roll_present = df[roll_col].notna().sum()
        roll_missing = df[roll_col].isna().sum()
    else:
        roll_present = "N/A"
        roll_missing = "N/A"
    
    results.append([name, total, aadhar_present, aadhar_missing, roll_present, roll_missing])

summary_df = pd.DataFrame(results, columns=[
    "Dataset", "Total Rows", "Aadhar Present", "Aadhar Missing", "Roll No Present", "Roll No Missing"
])

summary_df
```

```python
results = []

for name, df in datasets.items():
    total = df.shape[0]
    unique_aadhar = df["aadhar_no"].nunique(dropna=True) if "aadhar_no" in df.columns else None
    
    # handle roll number variations
    roll_col = None
    for candidate in ["roll_no", "RollNo"]:
        if candidate in df.columns:
            roll_col = candidate
            break
    
    unique_roll = df[roll_col].nunique(dropna=True) if roll_col else "N/A"
    
    # duplicates based on aadhar
    duplicates = total - unique_aadhar if unique_aadhar is not None else "N/A"
    
    results.append([name, total, unique_aadhar, unique_roll, duplicates])

unique_summary = pd.DataFrame(results, columns=[
    "Dataset", "Total Rows", "Unique Aadhar", "Unique Roll No", "Duplicate Rows"
])

unique_summary
```
